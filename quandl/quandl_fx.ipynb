```python
def get_pst_time():
    date_format='%Y-%m-%d %H:%M:%S %z'
    date = datetime.now(tz=pytz.utc)
    date = date.astimezone(timezone('US/Pacific'))
    pstDateTime=date.strftime(date_format)
    pstDateTime = "{0}:{1}".format(pstDateTime[:-2],pstDateTime[-2:])
    return pstDateTime
```
